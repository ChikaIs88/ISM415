```python
import pandas as pd
```

```python
data = pd.read_csv('NYC_Property_Data.csv')
```

```python
data.columns
```

```python
data
```

```python
data.isnull().sum
```

```python
data.dropna(how='any', inplace=True)
```

```python
data.isnull().sum
```

```python
data
```

```python
features = data.drop(['SALE PRICE', 'ADDRESS', 'SALE DATE'], axis=1)
```

```python
features
```

```python
task = data['SALE PRICE']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
features['NEIGHBORHOOD'] = le.fit_transform(features['NEIGHBORHOOD'])
```

```python
features.columns
```

```python
features['BUILDING CLASS CATEGORY'] = le.fit_transform(features['BUILDING CLASS CATEGORY'])
```

```python
features['TAX CLASS AT PRESENT'] = le.fit_transform(features['TAX CLASS AT PRESENT'])
```

```python
features['BUILDING CLASS AT PRESENT'] = le.fit_transform(features['BUILDING CLASS AT PRESENT'])
features['BUILDING CLASS AT TIME OF SALE'] = le.fit_transform(features['BUILDING CLASS AT TIME OF SALE'])
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, task_train, task_test = train_test_split(features, task, test_size = 0.2)
```

```python
features_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
nyc_model = lr.fit(features_train, task_train)
```

```python
task_pred = nyc_model.predict(features_test)
```

```python
pd.DataFrame({'Actual': task_test, 'Predicted': task_pred, 'Error': task_test-task_pred})
```
