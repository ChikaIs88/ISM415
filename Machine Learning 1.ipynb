```python
import pandas as pd
```

```python
data = pd.read_csv('50_Startups.csv')
```

```python
data.head()
```

```python
data.isnull().sum
```

```python
data
```

```python
# Drops profit since we are looking for everything but profit
features = data.drop('Profit', axis=1)
```

```python
# when you just want to see a the profit data
target = data['Profit']
```

```python
target
```

# We are going to convert the states to numbers to carry out machine learning


```python

from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
#we are going to use this function to perform the encoding, changinging the catergorical variables into numeric
features['State'] = le.fit_transform(features['State'])
```

```python
features
```

## Next stage in ML is to divide to data to test data and train data

```python
from sklearn.model_selection import train_test_split
```

```python
#You need to assign data set to your splits
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size = 0.2)
```

```python
features_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#this learns from data to make predictions. it is an algorithm
lr = LinearRegression()
```

```python
# .fit goes through features, learns it and spies on the target
model = lr.fit(features_train, target_train)
```

```python
target_pred = model.predict(features_test)
```

```python
#create  table to compare the prdicted values with the originals

pd.DataFrame({'Actual': target_test, 'Predicted': target_pred, 'Error': target_test-target_pred})
```
